```python
import sys
import time as time
import numpy as np
import DNA 
import population as pop
```

```python
sys.setrecursionlimit(10 ** 9)
```

```python
mutation_rate = 0.8
gen_count = 1
```

```python
def run_algorithm(gen_count):
    print("Enter Queen Numbers (Greater than 3): ")
    queens = int(input())
    print("Enter Population Size: ")
    pop_size = int(input())
    pop_ = pop.Population(pop_size, mutation_rate, queens, 5)

    t_start = time.time()
    while not pop_.is_solution_has_been_found():
        gen_count += 1
        print("\nGeneration Number: ", gen_count)
        pop_.natural_selection()
    t_end = time.time()
    print("Generation Number: ", gen_count)
    print("Time = ", t_end - t_start)
```

```python
run_algorithm(gen_count)
```
